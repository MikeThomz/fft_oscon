```python
import numpy as np
import matplotlib.pylab as plt
import scipy.io.wavfile as aud 

%matplotlib inline
```

```python
sample_rate, a = aud.read('/Users/william/Dropbox/guitar_plucking.wav')
```

```python
print "The sampling rate is " + str(sample_rate) + " samples per second"
print "There are " + str(len(a)) + " samples in the recording"
```

```python
plt.subplot(211)
plt.plot(a[:10000,0])
plt.subplot(212)
plt.plot(a[:10000,1])
```

```python
import scipy.fftpack as fft
xk = fft.fft(a[:10000,0])
```

```python
print "There are {0} bins in the FFT".format(len(xk))
plt.plot(np.abs(xk))
```

```python
fs = sample_rate/1000.0      # Let's show frequences in kilohertz (kHz)      
nBins = 5000
xk = fft.fft(a[:10000,0],n = nBins)   
freq_axis = np.linspace(-fs/2,fs/2,num = nBins)
plt.plot(freq_axis,np.abs(fft.fftshift(xk)))
plt.xlim((-5,5))
plt.xlabel('Frequency (kHz)')
plt.ylabel('Energy')
```

```python
fs = sample_rate/1000.0      # Let's show frequences in kilohertz (kHz)      
nBins = 7000
xk = fft.fft(a[:10000,0],n = nBins)   
freq_axis = np.linspace(0,fs/2,num = nBins/2)
plt.plot(freq_axis,np.abs(xk[:nBins/2]))
plt.xlim((0,1))
plt.xlabel('Frequency (kHz)')
plt.ylabel('Energy')

# Find max elements

nTop = 10
max_freqs = np.argsort(np.abs(xk[:nBins/2]))

plt.plot(freq_axis[max_freqs[-nTop:]],np.abs(xk[max_freqs[-nTop:]]),'rx',lw=2.0)

print "Max frequency is {0} kHz".format(freq_axis[max_freqs[-1]])
```

```python
# Find the frquency of the later note

fs = sample_rate/1000.0      # Let's show frequences in kilohertz (kHz)      
nBins = 7000

xk2 = fft.fft(a[20000:,0],n = nBins)

freq_axis = np.linspace(0,fs/2,num = nBins/2)
plt.plot(freq_axis,np.abs(xk2[:nBins/2]))
plt.xlim((0,1))
plt.xlabel('Frequency (kHz)')
plt.ylabel('Energy')

# Find max elements

nTop = 10
max_freqs = np.argsort(np.abs(xk2[:nBins/2]))

plt.plot(freq_axis[max_freqs[-nTop:]],np.abs(xk2[max_freqs[-nTop:]]),'rx')

print "Max frequency is {0} kHz".format(freq_axis[max_freqs[-1]])
```

```python
nSlices = 100
slice_width = np.floor(len(a)/nSlices)
nBins = slice_width

maxDispFreq = 1.0    # kHz, since fs is defined in kHz
maxDispBin = np.floor(maxDispFreq/(fs/nBins))    # freq/bin = fs/nBins

spectrogram = np.ones((nSlices,nBins))
fig = plt.figure(figsize=(9,9))
for i in range(nSlices):
    slice_start = i*slice_width
    slice_stop = (i+1)*slice_width
    slice_fft = fft.fft(a[slice_start:slice_stop,0]*np.hamming(slice_width),n = nBins)

    spectrogram[i,:] = np.log10(np.abs(slice_fft))
    

plt.imshow(spectrogram[:,:maxDispBin],cmap="BuPu") 
plt.ylabel('Time')
plt.xlabel('Frequency bins')
plt.colorbar()
```

```python
nSlices = 100
percent_overlap = 1.10
slice_width = np.floor((len(a))/nSlices)
nBins = slice_width

# Maximum frequency to display
maxDispFreq = 1.0    # kHz, since fs is defined in kHz
# How many bins does this translate into?
maxDispBin = np.floor(maxDispFreq/(fs/nBins))    # freq/bin = fs/nBins

spectrogram = np.ones((nSlices,nBins))
fig = plt.figure(figsize=(9,9))
for i in range(nSlices):
    slice_start = np.max([0,i*slice_width + slice_width*(1 - percent_overlap)])
    slice_stop = (i+1)*slice_width
    overlapped_width = slice_stop - slice_start
    slice_fft = fft.fft(a[slice_start:slice_stop,0]*np.hamming(overlapped_width),n = nBins)

    spectrogram[i,:] = np.log10(np.abs(slice_fft))
    

plt.imshow(spectrogram[:,:maxDispBin],cmap="BuPu") 
plt.ylabel('Time')
plt.xlabel('Frequency bins')
plt.colorbar()
```
